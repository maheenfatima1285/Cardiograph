## Dataset Description
### the dataset contains all numeric(float) values 
#### 1.LB-baseline fetal heartrate(FHR)-represent fhr over a period(110-160bpm is normal)
#### 2.AC-acceleration in FHR-a sign of fetal well being(more acceleration=better condition)
#### 3.FM-fetal movements detected by monitor, indicates fetal activity(high value=healthy fetus)
#### 4.UC-uterine contraction which impact the FHR,measure frequency(important during labor)
#### 5.DL-late deceleration-heartrate drops after contraction due to oxygen deficiency(which may be a warning sign)
#### 6.DS-short deceleration-Refers to brief drops in fetal heart rate(less severe than DP)
#### 7.DP-prolonged deceleration-long lastiong drops in heartrate(serious condition)
#### 8.ASTV-Abnormal Short-Term Variability-how much time heartrate was abnormal(high astv=bad condition)
#### 9.MSTV-average short term variability-small changes in heartrate(high mstv=good condition)
#### 10.ALTV-Abnormal Long-Term Variability-how much time long pattern was abnormal(high altv=bad condition)
#### 11.MLTV-average Long-Term Variability-how much time long pattern was good(high mltv=good condition)
#### 12.Width-range of fhr(high width can be abnormal)
#### 13.Tendency-show overall trend or direction of fhr(-1=Decreasing trend-can be risky ,0=normal ,1=generally okay)
#### 14.NSP-fetal state(It tells how healthy the condition is based on all the measurements (LB, AC, ASTV, etc.))

### Conclusion

#### The cardiotocography dataset focuses on analyzing fetal health conditions using various medical attributes. The target variable NSP (Fetal State) represents the final classification of fetal health into three categories: Normal (1), Suspect (2), and Pathological (3), making it the most important variable for prediction.

#### From the exploratory data analysis, it is observed that ASTV (Abnormal Short-Term Variability) and ALTV (Abnormal Long-Term Variability) show the highest correlation with NSP. Both variables have a moderate positive relationship with the target, indicating that as abnormal variability increases, the fetal condition tends to worsen.

#### Additionally, ASTV and ALTV are directly proportional to NSP, meaning higher variability is associated with higher risk levels. In contrast, FM (Fetal Movement) has a weak relationship with NSP, suggesting that fetal movement has less influence on determining the fetal health condition compared to variability measures. FM also shows an inverse trend with NSP, where lower movement may be associated with poorer conditions.

#### Overall, the analysis indicates that variability-based features (ASTV and ALTV) are key indicators of fetal health, while FM plays a comparatively minor role. These insights can help in building predictive models and assist healthcare professionals in early detection and monitoring of fetal risk conditions
